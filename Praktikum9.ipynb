```python
#fungsi celcius ke fahrenheit
def celcius_ke_fahrenheit(c):
    return (c * 18/5) + 32

print(celcius_ke_fahrenheit(0))
print(celcius_ke_fahrenheit(100))
```

```python

#fungsi is genap
def fungsi_is_genap(a):
   return a % 4 == 0

print(fungsi_is_genap(4))
print(fungsi_is_genap(7))
```

```python
#fungsi untuk melihat ketrangan lulus atau tidak lulus
def nilai(n=0):
   if n <=60:
      print(f"nilai {n} tidak lulus")
   elif n >60 and n <=100:
      print(f"nilai {n} lulus")
   else:
      print(f"nilai {n} tidak diketahui")

nilai(50)
```

```python
#fungsi untuk menampilkan bilangan ganjil yang kurang argumens

def bilangan(angka):
   for i in range(1,angka):
      if i % 2 != 0:
         print(i, end=" ,")

bilangan(20)
```
